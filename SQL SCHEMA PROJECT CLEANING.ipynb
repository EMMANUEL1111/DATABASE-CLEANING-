```python
resturant_df['Email'].fillna('Not Available', inplace=True)
resturant_df['Address'].fillna('Unknown Address', inplace=True)
resturant_df['Manager'].fillna('Unknown Manager',inplace=True)
resturant_df['Years_as_manager'].fillna(value=resturant_df['Years_as_manager'].mean(),inplace=True)
```

```python
orders_df['Credit Card'].fillna('Not Provided', inplace=True)
orders_df['Debit Card'].fillna('Not Provided', inplace=True)
orders_df['Customer Rating-Food'].fillna(0, inplace=True)
orders_df['Customer Rating-Delivery'].fillna(0, inplace=True)
orders_df['Quantity of Items'].fillna(orders_df['Quantity of Items'].mean(),inplace=True)
orders_df['Payment Mode'].fillna("Not Available",inplace=True)
orders_df['Card provider'].fillna("Not Available",inplace=True)
orders_df['First_Customer_Name'].fillna('Not Provided',inplace=True)
```
